```python
#Final Test Dimbimbing Fathur Rahman Syahputra

#1 Buatlah API Request
import requests
import pandas as pd

url = 'https://berita-indo-api-next.vercel.app/api/cnn-news/teknologi'
response = requests.get(url)
data = response.json()['data']

news_data = []

for item in data:
  if 'akcdn.detik.net.id' in item['link']:
    creator = 'detik'
  elif 'www.cnnindonesia.com' in item['link']:
    creator = 'cnn'
  else:
    creator = '-'

  news = {
    'creator': creator,
    'title': item['title'],
    'link': item['link'],
    'categories': 'teknologi',
    'isodate': item['isoDate'],
    'description': item['contentSnippet'],
    'image': item['image']['small']
  }

  news_data.append(news)

#2 Transformasi menjadi dataframe
df = pd.DataFrame(news_data)
df
```

```python
#4 Transformasi data kolom isodate menjadi tipe data datetime
df['isodate'] = pd.to_datetime(df['isodate'])

df.dtypes
```

```python
#5 Aggregasi berdasarkan kolom 'creator', serta urutkan jumlah terbanyak
creator_counts = df.groupby('creator').size().reset_index(name='total')
creator_counts = creator_counts.sort_values('total', ascending=False)

creator_counts
```
